## Tranning The Model By The Data Set And We Use Random Forest Classifier

```python
x = df.drop(['AQI_Category','CO_AQI_Category','Ozone_AQI_Category','NO2_AQI_Category','PM_25_AQI_Category'],axis=1); # Independent Variable (Input)
y = df[['AQI_Category','CO_AQI_Category','Ozone_AQI_Category','NO2_AQI_Category','PM_25_AQI_Category']] # Dependent Variable (Output)
```

```python
x
```

```python
y
```

#### Now We Split The Data Into Two Parts I.E Train Data And Test Data

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
# Here The 42 Is Used For Constant Data Spliting When We Rerun The Code. 
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.multioutput import MultiOutputClassifier

model = MultiOutputClassifier(RandomForestClassifier())

model.fit(x_train, y_train) # Helps To Train The Data. 
```

```python
print(f"The Score Of Train Data Is: ",model.score(x_train,y_train))
print(f"The Score Of Test Data Is: ",model.score(x_test,y_test))
```

```python
y_pred = model.predict(x_test); # Jo Tumne Data Test Ke lia Lia Tha Yani Input Data (Independent) Usse Test kar Rahe Hai
y_pred
```

```python
import joblib
joblib.dump(model, 'AQI_Model.joblib')
```
