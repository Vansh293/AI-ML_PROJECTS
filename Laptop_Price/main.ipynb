```python
unq_brand = df['brand'].unique()
unq_processor_brand = df['processor_brand'].unique()
unq_processor_name = df['processor_name'].unique()
unq_processor_gnrtn = df['processor_gnrtn'].unique()
unq_ram_gb = df['ram_gb'].unique()
unq_ssd = df['ssd'].unique()
unq_hdd = df['hdd'].unique()
unq_os = df['os'].unique()
unq_os_bit = df['os_bit'].unique()
unq_graphic_card_gb = df['graphic_card_gb'].unique()
unq_weight = df['weight'].unique()
unq_warranty = df['warranty'].unique()
unq_touchscreen = df['Touchscreen'].unique()
unq_rating = df['rating'].unique()
```

```python
dict_brand = {val:i+1 for i,val in enumerate(unq_brand)}
dict_processor_brand = {val: i+1 for i, val in enumerate(unq_processor_brand)}
dict_processor_name = {val: i+1 for i, val in enumerate(unq_processor_name)}
dict_processor_gnrtn = {val: i+1 for i, val in enumerate(unq_processor_gnrtn)}
dict_ram_gb = {val: i+1 for i, val in enumerate(unq_ram_gb)}
dict_ssd = {val: i+1 for i, val in enumerate(unq_ssd)}
dict_hdd = {val: i+1 for i, val in enumerate(unq_hdd)}
dict_os = {val: i+1 for i, val in enumerate(unq_os)}
dict_os_bit = {val: i+1 for i, val in enumerate(unq_os_bit)}
dict_graphic_card_gb = {val: i+1 for i, val in enumerate(unq_graphic_card_gb)}
dict_weight = {val: i+1 for i, val in enumerate(unq_weight)}
dict_warranty = {val: i+1 for i, val in enumerate(unq_warranty)}
dict_touchscreen = {val: i+1 for i, val in enumerate(unq_touchscreen)}
dict_rating = {val: i+1 for i, val in enumerate(unq_rating)}
```

```python
df['brand'] = df['brand'].map(dict_brand)
df['processor_brand'] = df['processor_brand'].map(dict_processor_brand)
df['processor_name'] = df['processor_name'].map(dict_processor_name)
df['processor_gnrtn'] = df['processor_gnrtn'].map(dict_processor_gnrtn)
df['ram_gb'] = df['ram_gb'].map(dict_ram_gb)
df['ssd'] = df['ssd'].map(dict_ssd)
df['hdd'] = df['hdd'].map(dict_hdd)
df['os'] = df['os'].map(dict_os)
df['os_bit'] = df['os_bit'].map(dict_os_bit)
df['graphic_card_gb'] = df['graphic_card_gb'].map(dict_graphic_card_gb)
df['weight'] = df['weight'].map(dict_weight)
df['warranty'] = df['warranty'].map(dict_warranty)
df['Touchscreen'] = df['Touchscreen'].map(dict_touchscreen)
df['rating'] = df['rating'].map(dict_rating)

```

```python
df
```

```python
x = df.drop(['Price'],axis=1) # Input
y = df['Price'] # Output
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor(n_estimators=100, random_state=42)

model.fit(x_train, y_train)
```

```python
print("The Train Score Is: ",model.score(x_train,y_train))
print("The Test Score Is: ",model.score(x_test,y_test))
```

```python
y_pred = model.predict(x_test)
y_pred
```
