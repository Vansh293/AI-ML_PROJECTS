```python
unq_sex = df['sex'].unique()
unq_sex

dict_sex = {
        'female':1, 
        'male':2
}
df['sex'] = df['sex'].map(dict_sex)
```

```python
unq_region = df['region'].unique()
unq_region

dict_region = {
    'southwest':1, 'southeast':2, 'northwest':3, 'northeast':4
}
df['region'] = df['region'].map(dict_region)
```

```python
unq_smoker = df['smoker'].unique()
unq_smoker

dict_smoker = {
    'yes':1, 'no':2
}

df['smoker'] = df['smoker'].map(dict_smoker)
```

```python
df
```

### ENCODED SUCCESSFULLY NOW TRAIN THE MODEL
---

```python
x = df.drop(['charges'],axis=1) # independent Variable (Input)
y = df['charges']
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

### HERE WE PREDICT THE VALUES SO THAT USE RANDOM FOREST REGRESSOR

```python
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor(n_estimators=100, random_state=42)

model.fit(x_train, y_train)
```

```python
print("The Train Score Is: ",model.score(x_train,y_train))
print("The Test Score Is: ",model.score(x_test,y_test))
```

```python
y_pred = model.predict(x_test)
y_pred
```
